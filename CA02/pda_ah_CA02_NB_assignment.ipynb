## Function: make_Dictionary(root_dir)

a)	Inputs: root_dir (path to the training email folder).
b)	Lists all files in root_dir and builds full paths using os.path.join.
c)	Reads every email file line-by-line and splits each line into tokens using whitespace split().
d)	Aggregates all tokens from all emails into one long list (all_words).
e)	Uses Counter(all_words) to count word frequencies across the entire training corpus.
f)	Cleans the vocabulary by deleting tokens that are not purely alphabetic (item.isalpha() == False) and tokens with length 1.
g)	Keeps only the top 3000 most common tokens using most_common(3000).
h)	Returns a list of 3000 (word, frequency) tuples. This list is the vocabulary used later for feature extraction.


```python
def make_Dictionary(root_dir):
  all_words = []
  emails = [os.path.join(root_dir,f) for f in os.listdir(root_dir)]
  for mail in emails:
    with open(mail) as m:
      for line in m:
        words = line.split()
        all_words += words
  dictionary = Counter(all_words)
  list_to_remove = list(dictionary)

  for item in list_to_remove:
    if item.isalpha() == False:
      del dictionary[item]
    elif len(item) == 1:
      del dictionary[item]
  dictionary = dictionary.most_common(3000)
  return dictionary
```

# Import all other necessary libraries


```python
from pathlib import Path
import re
from collections import Counter

import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# call from local folders
# TRAIN_DIR = Path("./train-mails")
# TEST_DIR  = Path("./test-mails")

MAX_FEATURES = 3000   # top-K words
MIN_TOKEN_LEN = 2     # drop very short tokens
```

# Data Reading & Labeling

We load emails from a directory and assign labels using the file-naming convention:
- Spam files begin with **"spmsg"** → label = 1
- Non-spam files begin with other prefix (e.g., "3-1msg1.txt") → label = 0

Improvements:
- Uses **Path.glob** + **sorted** for consistent ordering.
- Uses encoding-safe reads (latin-1 fallback).


```python
#def list_email_files(folder: Path):
#    """Return a reproducible sorted list of email file paths."""
#    return sorted([p for p in folder.iterdir() if p.is_file()])

def list_email_files(folder):
    folder = Path(folder)   # auto-convert if string
    return sorted([p for p in folder.iterdir() if p.is_file()])


def infer_label_from_filename(file_path: Path) -> int:
    """Label rule: spam if filename starts with 'spmsg'."""
    return 1 if file_path.name.startswith("spmsg") else 0

def read_email_text(file_path: Path) -> str:
    """Read full email text safely (avoid decoding crashes)."""
    try:
        return file_path.read_text(encoding="utf-8", errors="ignore")
    except Exception:
        return file_path.read_text(encoding="latin-1", errors="ignore")
```

# Tokenization (Basic Normalization)

Original code used `split()` and then removed non-alphabetic tokens.
This improved version:
- Lowercases text (so 'Free' == 'free')
- Uses regex to extract tokens
- Keeps tokens that are alphabetic, but you can choose to include digits if desired

Note:
- You mentioned that spam signals can include digits/URLs.
- If you want to keep numbers, we can adjust the regex to include them.


```python
TOKEN_PATTERN = re.compile(r"[a-zA-Z]+")

def tokenize(text: str):
    """Convert raw email text into a list of normalized tokens."""
    text = text.lower()
    tokens = TOKEN_PATTERN.findall(text)   # alphabetic tokens only
    tokens = [t for t in tokens if len(t) >= MIN_TOKEN_LEN]
    return tokens
```

# Vocabulary Builder (Feature Dictionary)

We build a dictionary from TRAIN emails only:
- Read each email
- Tokenize
- Count frequencies across corpus
- Keep top `MAX_FEATURES` tokens

Improvements:
- Uses `Counter.update()` for O(n) counting
- Produces both:
  - `vocab_list` = list of top words
  - `word_to_index` = fast lookup mapping (O(1))


```python
def build_vocabulary(train_folder: Path, max_features=3000):
    corpus_counter = Counter()

    for fp in list_email_files(train_folder):
        text = read_email_text(fp)
        tokens = tokenize(text)
        corpus_counter.update(tokens)

    most_common = corpus_counter.most_common(max_features)
    vocab_list = [w for w, _ in most_common]
    word_to_index = {w: i for i, w in enumerate(vocab_list)}

    return vocab_list, word_to_index, most_common
```

# Feature Extraction (Sparse Bag-of-Words)

We convert emails to numeric features using the vocabulary:
- Each email becomes a vector length = 3000
- Vector value = count of each vocab word in email

Improvements:
- Uses Counter once per email (no O(n^2) `words.count()` loops)
- Uses a sparse matrix (CSR) for memory efficiency


```python
def featurize_emails(folder: Path, word_to_index: dict, vocab_size: int):
    files = list_email_files(folder)

    rows, cols, data = [], [], []
    y = np.zeros(len(files), dtype=int)

    for doc_id, fp in enumerate(files):
        # label
        y[doc_id] = infer_label_from_filename(fp)

        # text -> tokens -> counts
        text = read_email_text(fp)
        tokens = tokenize(text)
        token_counts = Counter(tokens)

        # build sparse row
        for token, cnt in token_counts.items():
            if token in word_to_index:
                rows.append(doc_id)
                cols.append(word_to_index[token])
                data.append(cnt)

    X = csr_matrix((data, (rows, cols)), shape=(len(files), vocab_size), dtype=np.float64)
    return X, y, files
```

# Train & Evaluate Naive Bayes

We train Multinomial Naive Bayes (standard for word-count features):
- Fit on TRAIN sparse feature matrix
- Predict on TEST
- Evaluate using accuracy + confusion matrix + classification report


```python
# 1) Build vocabulary from TRAIN
vocab_list, word_to_index, vocab_with_counts = build_vocabulary(TRAIN_DIR, MAX_FEATURES)
print(f"Vocabulary size: {len(vocab_list)}")

# 2) Featurize TRAIN and TEST using same vocabulary
X_train, y_train, train_files = featurize_emails(TRAIN_DIR, word_to_index, len(vocab_list))
X_test,  y_test,  test_files  = featurize_emails(TEST_DIR,  word_to_index, len(vocab_list))

print("Train matrix:", X_train.shape, "Test matrix:", X_test.shape)

# 3) Train NB
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)

# 4) Predict + evaluate
y_pred = nb_model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
cm  = confusion_matrix(y_test, y_pred)

print("Accuracy:", round(acc, 4))
print("Confusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=4))
```
