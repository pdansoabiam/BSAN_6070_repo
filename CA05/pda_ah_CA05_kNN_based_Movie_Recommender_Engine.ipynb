#CA05 - kNN based Movie Recommender Engine

A k-Nearest Neighbors recommender system built with Python and scikit-learn that finds the 5 most similar movies to a given query movie based on genre and IMDB rating data.

```python
# import libraries

import pandas as pd
import numpy as np

from sklearn.neighbors import NearestNeighbors
```

Load Data

```python
# Data source provided in instructions

data_url = "https://raw.githubusercontent.com/ArinB/MSBA-CA-Data/main/CA05/movies_recommendation_data.csv"

# Read directly from GitHub
movies_df = pd.read_csv(data_url)

```

Explore the Data

```python
# Check dataset shape
print("Dataset shape:", movies_df.shape)

print(movies_df.columns.tolist())

# Display first 5 rows
movies_df.head()
```

 Prepare Features

```python
# Keep the numeric movie features used by the recommender
feature_cols = [
    'IMDB Rating', 'Biography', 'Drama', 'Thriller',
    'Comedy', 'Crime', 'Mystery', 'History'
]

X = movies_df[feature_cols]

X.head()
```

 Build the kNN Model

```python
# Fit the kNN model
# We use 5 neighbors because 'The Post' is not part of the dataset.

neigh = NearestNeighbors(n_neighbors=5, metric='euclidean')
neigh.fit(X)
```

Define "The Post" Query Vector

```python
# Feature vector for 'The Post'
# IMDB Rating=7.2, Biography=1, Drama=1, Thriller=0,
# Comedy=0, Crime=0, Mystery=0, History=1

the_post = pd.DataFrame([[7.2, 1, 1, 0, 0, 0, 0, 1]], columns=feature_cols)

the_post
```

 Find 5 Most Similar Movies

```python
# kneighbors returns (distances, indices)

distances, indices = neigh.kneighbors(the_post)

print("Indices :", indices)
print("Distances:", distances)
```

Display Recommendations

```python
# Display the 5 recommended movies

recommended_movies = movies_df.iloc[indices[0]]['Movie Name'].reset_index(drop=True)
recommended_distances = distances[0]

print("Top 5 Movies Similar to 'The Post':\n")

for i, (movie, distance) in enumerate(zip(recommended_movies, recommended_distances), start=1):
    print(f"{i}. {movie} - Distance: {distance:.4f}")
```

Clean Final Output

```python
# Final answer table

result_df = pd.DataFrame({
    'Rank': range(1, 6),
    'Movie Name': recommended_movies.values,
    'Distance': recommended_distances
})

result_df
```

The kNN recommender was built successfully and identified the 5 movies most similar to "The Post" based on IMDB rating and genre features: 12 Years a Slave, Hacksaw Ridge, Queen of Katwe, The Wind Rises, and A Beautiful Mind.
